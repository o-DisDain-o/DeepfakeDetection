```python
from facenet_pytorch import MTCNN
import cv2
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
from tqdm.notebook import tqdm
```

```python
import json
import csv
# Read the JSON file
with open('./train_sample_videos/metadata.json', 'r') as file:
    metadata = json.load(file)
```

```python
from alive_progress import alive_bar
# import cv2
# from PIL import Image
# from mtcnn import MTCNN
# import csv

def mtcnn_func(videoname, writer):
    # Create face detector
    mtcnn = MTCNN(margin=20, keep_all=True, post_process=False)

    # Load a video
    v_cap = cv2.VideoCapture("./train_sample_videos/" + videoname)
    v_len = int(v_cap.get(cv2.CAP_PROP_FRAME_COUNT))
    height = v_cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
    width = v_cap.get(cv2.CAP_PROP_FRAME_WIDTH)

    # Loop through video, taking a handful of frames to form a batch
    frames = []
    with alive_bar(v_len, title='Processing') as bar:
        for i in range(v_len):
            # Load frame
            success = v_cap.grab()
            if i % 50 == 0:
                success, frame = v_cap.retrieve()
            else:
                continue
            if not success:
                continue
            
            # Add to batch
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(Image.fromarray(frame))
            
            bar()

    # Detect faces in batch
    save_paths = [f'./DFDC_Images/{videoname[:-4]}_{i}.jpg' for i in range(len(frames))]
    
    for i, frame in enumerate(frames):
        writer.writerow([f'{videoname[:-4]}_{i}.mp4', width, height, value["label"], value["original"]])
    
    faces = mtcnn(frames, save_path=save_paths)
```

```python
i = 0
videoPath = "./train_sample_videos/"

with open('DFDC_metadata.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    field = ["videoname", "original_width", "original_height", "label", "original"]
    writer.writerow(field)
    
    for key, value in metadata.items():
        if value["label"] == "FAKE":
            isReal = False
        else :
            isReal = True

        videoName = key
        mtcnn_func(videoName, writer)
```
